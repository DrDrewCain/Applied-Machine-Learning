```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# How many features/attributes are there per sample? How many different
There are total of 4 different features per sample: sepal length, sepal width, petal length, and petal width, whcih are all measured in centimeters.
# species are there, and how many samples of each species did Anderson record?
There are a total of 150 samples based on (50 samples) per species. (species being Iris Setosa, Iris Versicolour, and Iris Virginica)

```python
# load the data into a pandas dataframe
df = pd.read_csv('./iris/iris.data', header=None)

# extract the features and labels
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values

# create an N-dimensional vector containing each sample's label
labels = np.array(y)
```

```python
# create a dictionary that maps each label to a unique color
colors = {'Iris-setosa': 'r', 'Iris-versicolor': 'g', 'Iris-virginica': 'b'}

# builds an N-dimensional vector containing each sample's label
c_values = [colors[label] for label in y]

# attribute names as shown in iris.names
attributes = ['sepal length in cm', 'sepal width in cm', 'petal length in cm', 'petal width in cm']

# create a grid of subplots
fig, axes = plt.subplots(4, 4, figsize=(12, 12))

# Iterate through all pairs of attributes
p = X.shape[1]
for i in range(p):
    for j in range(p):
        if i != j:
            axes[i, j].scatter(X[:, i], X[:, j], c=c_values)
            axes[i, j].set_xlabel(attributes[i])
            axes[i, j].set_ylabel(attributes[j])
        else:
            axes[i, j].text(0.5, 0.5, attributes[i], ha='center')

plt.tight_layout()
plt.show()
```
